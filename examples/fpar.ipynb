```python
import numpy as np
import matplotlib.pyplot as plt
from fse_toolkit.utils.common import read_tif
from fse_toolkit.multispectral.fpar import calculate_ndvi, calculate_sr, calculate_fpar
```

```python
tif_files = [
    '../../../FPAR/To彥成/95183018_221211e_28~6584_hr4-003.tif',
    '../../../FPAR/To彥成/95183020_221120f_30~0012_hr4.tif',
    '../../../FPAR/To彥成/95183029_221120f_29~0113_hr4.tif',
    '../../../FPAR/To彥成/95183038_221211e_28~6592_hr4-019.tif',
    '../../../FPAR/To彥成/95183040_221120f_30~0004_hr4.tif',
    '../../../FPAR/To彥成/95183049_221120f_29~0121_hr4.tif'
]
```

```python
tif_data = read_tif(tif_files[1])
```

```python
ndvi = calculate_ndvi(tif_data["bands"][3], tif_data["bands"][0]) # calculate NDVI
NDVI_min = np.nanmin(ndvi)
NDVI_max = np.nanmax(ndvi)
print(f"NDVI 最小值: {NDVI_min}, 最大值: {NDVI_max}")
```

```python
# show ndvi_filtered by matplotlib
plt.imshow(ndvi, cmap='RdYlGn')
plt.show()
```

```python
sr = calculate_sr(ndvi)
SR_min = np.nanmin(sr)
SR_max = np.nanmax(sr)
print(f"SR 最小值: {SR_min}, 最大值: {SR_max}")
```

```python
plt.imshow(sr, cmap='RdYlGn')
plt.show()
```

```python
fpar = calculate_fpar(ndvi, sr)
FPAR_min = np.nanmin(fpar)
FPAR_max = np.nanmax(fpar)
print(f"FPAR 最小值: {FPAR_min}, 最大值: {FPAR_max}")
```

```python
plt.imshow(fpar, cmap='RdYlGn')
plt.show()
```
